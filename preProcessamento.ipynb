```python
import pandas as pd
import re

import numpy as np# read csv data
data = pd.read_csv(r'D:\roger\Documents\Incor\EVOLUCAO-AMOSTRA.csv', encoding='latin1', sep=';')
data.head()
```

```python
file = open("textoProcessado.txt", "w+") # Abre o arquivo de texto


for texto in data.TEXTO_ATENDIMENTO_ANONIMIZADO:
    textoProcessado = re.sub(r'[#]',' ', texto) # Remove todos os # e substitui por nada
    textoProcessado = re.sub(r'\n\n',' ', textoProcessado)
    textoProcessado = re.sub(r'[\n]','. ', textoProcessado) # Remove todos as quebras de linha e substitui por nada
    textoProcessado = re.sub(r'[:]',' ', textoProcessado) # Remove todos os : e substitui por nada
    textoProcessado = re.sub(r'[-]',' ', textoProcessado) # Remove todos os - e substitui por nada
    textoProcessado = re.sub(r'//',' ', textoProcessado) # Remove todos os // e substitui por nada
    textoProcessado = re.sub(r'[;]','. ', textoProcessado) # Remove todos os ; e substitui por nada
    textoProcessado = re.sub(r'[< >]',' ', textoProcessado) # Remove todos os < ou > e substitui por nada
    #textoProcessado = re.sub(r'. .','', textoProcessado) # Remove todos os . . e substitui por .
    #textoProcessado = re.sub(r'..','.', textoProcessado) # Remove todos os .. e substitui por .
    #textoProcessado = re.sub(r':.','.', textoProcessado) # Remove todos os :. e substitui por .
    #textoProcessado = re.sub(r': .','.', textoProcessado) # Remove todos os : . e substitui por .
    #Remove somente se encontrar IDENTIFICACAO DO PACIENTE E uma sequência de 8 números de 0 a 9 
    textoProcessado = re.sub(r'IDENTIFICACAO DO PACIENTE + [0-9]{8}',' ', textoProcessado)
    #Remove somente se encontrar SEQUENCIA DO EVENTO E uma sequência de 2 números de 0 a 9 
    textoProcessado = re.sub(r'SEQUENCIA DO EVENTO + [0-9]{2}',' ', textoProcessado)
    #Remove somente se encontrar SEQUENCIA DO EVENTO E uma sequência de 1 número de 0 a 9 
    textoProcessado = re.sub(r'SEQUENCIA DO EVENTO + [0-9]{1}',' ', textoProcessado)
    #Remove somente se encontrar INTERVALO EM DIAS ENTRE OS EVENTOS E uma sequência de 3 números de 0 a 9
    textoProcessado = re.sub(r'INTERVALO EM DIAS ENTRE OS EVENTOS + [0-9]{3}',' ', textoProcessado)
    #Remove somente se encontrar INTERVALO EM DIAS ENTRE OS EVENTOS E uma sequência de 2 números de 0 a 9
    textoProcessado = re.sub(r'INTERVALO EM DIAS ENTRE OS EVENTOS + [0-9]{2}',' ', textoProcessado)
    #Remove somente se encontrar INTERVALO EM DIAS ENTRE OS EVENTOS E uma sequência de 1 número de 0 a 9 
    textoProcessado = re.sub(r'INTERVALO EM DIAS ENTRE OS EVENTOS + [0-9]{1}',' ', textoProcessado)
    textoProcessado = " ".join(textoProcessado.split()) # Remove os espaços sobrando
    textoProcessado = ".".join(textoProcessado.split('. .'))
    textoProcessado = ".".join(textoProcessado.split('..'))
    textoProcessado = ".".join(textoProcessado.split(':.'))
    textoProcessado = ".".join(textoProcessado.split(': .'))
    
    print(textoProcessado)
    file.write(textoProcessado)

file.close()   
```
